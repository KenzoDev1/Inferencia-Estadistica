5.2. Usa la variable `promedio_final`.

a) La variable que se describe, ¿se distribuye normalmente? ¿Por qué? Justifique analizando las medidas de distribución, el gráfico correspondiente y las medidas de tendencia central.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("G0_A2_estudiantes.xlsx")

# Para saber si la variable "promedio_final" se distribuye normalmente
# Hay que comprobar varias cosas

# 1 Medidas de tendencia central
media = df["promedio_final"].mean()
mediana = df["promedio_final"].median()
moda = df["promedio_final"].mode()[0]

print("Media:", round(media,2))
print("Mediana:", round(mediana,2))
print("Moda:", round(moda,2))
# Si la media, mediana y moda son similares, es un indicio de que la distribución es simétrica 
# y podría ser normal.

# 2 Medidas de distribucion
asimetria = df["promedio_final"].skew()
curtosis = df["promedio_final"].kurt()

print("Asimetría:", round(asimetria,2)) # Que tan ladeada esta la curva, asimetria negativa
print("Curtosis:", round(curtosis,2)) # Que tan puntiaguda es la curva
# Si la asimetria esta entre -0.5 y 0.5, se considera que la distribución
# es aproximadamente simétrica.
# Si la curtosis esta entre -1 y 1, se considera que la distribución 
# es aproximadamente mesocúrtica, lo que es típico de una distribución normal.

# 3 Analisis grafico
plt.figure(figsize=(10, 6))
plt.hist(df["promedio_final"], bins=20, edgecolor='black', color='steelblue')
plt.xlabel('Promedio Final')
plt.ylabel('Frecuencia')
plt.title('Distribución del Promedio Final', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.show()

# Segun esto se puede observar que la distribución del promedio final es aproximadamente normal, 
# ya que la media, mediana y moda son similares, 
# la asimetria esta cerca de 0 y la curtosis esta cerca de 0.
# y como ultimo en el histograma se ve que no hay nada fuera de lo normal, no hay
# ladeadas en la curva ni barras muy pronunciadas.
```

6.2. Asumiendo que la variable `horas_estudio` distribuye aproximadamente normal, determine su media y desviación estándar. Luego, responda las preguntas indicadas.

```python
media = df["horas_estudio"].mean()
desviacion_estandar = df["horas_estudio"].std()
print("La media de las horas de estudio es:", round(media,2))
print("La desviación estándar de las horas de estudio es:", round(desviacion_estandar,2))
```

a) ¿Cuál es la probabilidad de que un estudiante dedique menos de 10 horas a estudiar?

```python
from scipy.stats import norm
prob_a = norm.cdf(10, media, desviacion_estandar)
print(f"La probabilidad es: {round(prob_a * 100, 2)}%")
```

b) ¿Cuál es la probabildiad de que un estudiante dedique entre 15 y 20 horas a estudiar?

```python
prob_b = norm.cdf(20, media, desviacion_estandar) - norm.cdf(15, media, desviacion_estandar)
print(f"La probabilidad entre 15 y 20 es: {round(prob_b * 100, 2)}%")
```

c) ¿Cuál es la probabilidad de que un estudiante dedique más de 11 horas a estudiar?

```python
prob_C = norm.sf(11, media, desviacion_estandar)
print(f"La probabilidad de que las horas de estudio sean mayores a 11 es: {round(prob_C * 100, 2)}%")
```
